```python
#1-What is the average listing price by city?
query = """
SELECT 
    City,
    ROUND(AVG(price), 2) AS avg_listing_price
FROM listings
WHERE price IS NOT NULL
GROUP BY City
ORDER BY avg_listing_price DESC;
"""

# Execute query
result_df = pd.read_sql(query, engine)

# Show output
result_df.head()
```

```python
#2-What is the average price per square foot by property type?
query = """
SELECT 
    property_type,
    ROUND(AVG(price / sqft), 2) AS avg_price_per_sqft
FROM listings
WHERE price IS NOT NULL 
  AND sqft IS NOT NULL
  AND sqft > 0
GROUP BY property_type
ORDER BY avg_price_per_sqft DESC;
"""

result_df = pd.read_sql(query, engine)

result_df.head()
```

```python
#3-How does furnishing status impact property prices?
query = """
SELECT 
    property.furnishing_status,
    AVG(listings.Price) AS avg_price
FROM listings
JOIN property
    ON listings.Listing_ID = property.listing_id
GROUP BY property.furnishing_status;
"""

result_df = pd.read_sql(query, engine)

result_df.head()
```

```python
#4-Do properties closer to metro stations command higher prices?

query = """
SELECT 
    CASE 
        WHEN property.metro_distance_km <= 1 THEN 'Near Metro'
        ELSE 'Far from Metro'
    END AS near_metro,
    AVG(listings.Price) AS avg_price
FROM listings
JOIN property
    ON listings.Listing_ID = property.listing_id
GROUP BY 
    CASE 
        WHEN property.metro_distance_km <= 1 THEN 'Near Metro'
        ELSE 'Far from Metro'
    END;
    """

result_df = pd.read_sql(query, engine)

result_df.head()
```

```python
#5-Are rented properties priced differently from non-rented ones?
query = """
SELECT 
    CASE 
        WHEN property.is_rented = 1 THEN 'Rented'
        ELSE 'Not Rented'
    END AS rental_status,
    AVG(listings.Price) AS avg_price
FROM listings
JOIN property
    ON listings.Listing_ID = property.listing_id
GROUP BY property.is_rented;
"""
result_df = pd.read_sql(query, engine)

result_df.head()
```

```python
#6-How do bedrooms and bathrooms affect pricing?
query = """
SELECT 
    property.bedrooms,
    property.bathrooms,
    AVG(listings.Price) AS avg_price
FROM listings
JOIN property
    ON listings.Listing_ID = property.listing_id
GROUP BY property.bedrooms, property.bathrooms;
"""
result_df = pd.read_sql(query,engine)
result_df.head()
```

```python
#7-Do properties with parking and power backup sell at higher prices?
query = """
SELECT 
    property.parking_available,
    property.power_backup,
    AVG(listings.Price) AS avg_price
FROM listings
JOIN property
    ON listings.Listing_ID = property.listing_id
GROUP BY property.parking_available, property.power_backup;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#8- How does year built influence listing price?
query = """
SELECT 
    property.year_built,
    AVG(listings.Price) AS avg_price
FROM listings
JOIN property
    ON listings.Listing_ID = property.listing_id
GROUP BY property.year_built;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#9-Which cities have the highest average property prices?
query = """
SELECT 
    listings.City,
    AVG(listings.Price) AS avg_price
FROM listings
GROUP BY listings.City
ORDER BY avg_price DESC;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#10- How are properties distributed across price buckets?
query = """
SELECT 
    CASE 
        WHEN listings.Price <= 3000000 THEN '0-30L'
        WHEN listings.Price <= 5000000 THEN '30L-50L'
        WHEN listings.Price <= 7000000 THEN '50L-70L'
        WHEN listings.Price <= 9000000 THEN '70L-90L'
        ELSE '90L+'
    END AS price_range,
    COUNT(*) AS property_count
FROM listings
GROUP BY 
    CASE 
        WHEN listings.Price <= 3000000 THEN '0-30L'
        WHEN listings.Price <= 5000000 THEN '30L-50L'
        WHEN listings.Price <= 7000000 THEN '50L-70L'
        WHEN listings.Price <= 9000000 THEN '70L-90L'
        ELSE '90L+'
    END
ORDER BY price_range;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#11-What is the average days on market by city?
query = """
SELECT 
    listings.City,
    AVG(sales.Days_on_Market) AS avg_days_on_market
FROM sales
JOIN listings
    ON sales.Listing_ID = listings.Listing_ID
GROUP BY listings.City;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#12-Which property types sell the fastest?
query = """
SELECT
listings.property_Type,
AVG(sales.Days_on_Market) AS avg_days_on_market
FROM Sales
JOIN Listings
ON Sales.Listing_ID = listings.Listing_ID
GROUP BY listings.Property_Type
ORDER BY avg_days_on_market ASC;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#13-What percentage of properties are sold above listing price?
query = """
SELECT 
(SUM(CASE 
WHEN sales.Sale_Price > listings.Price THEN 1 
ELSE 0 
 END) * 100.0) / COUNT(*) AS percentage_higher_than_list_price
FROM sales
JOIN listings
ON sales.Listing_ID = listings.Listing_ID;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
    
```

```python
#14-What is the sale-to-list price ratio by city?
query = """
SELECT 
listings.City,
AVG(sales.Sale_Price / listings.Price) * 100 AS avg_price_ratio_percentage
FROM sales
JOIN listings
ON sales.Listing_ID = listings.Listing_ID
GROUP BY listings.City;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#15-What is the sale-to-list price ratio by city?
query = """
SELECT 
    listings.City,
    (AVG(CASE 
        WHEN sales.Days_on_Market > 90 THEN 1 
        ELSE 0 
    END) * 100) AS percentage_above_90_days
FROM sales
JOIN listings
    ON sales.Listing_ID = listings.Listing_ID
GROUP BY listings.City;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#16-Which listings took more than 90 days to sell?
query = """
SELECT *
FROM sales
JOIN listings
    ON sales.Listing_ID = listings.Listing_ID
WHERE sales.Days_on_Market > 90;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#17-How does metro distance affect time on market?
query = """
SELECT 
    property.metro_distance_km,
    AVG(sales.Days_on_Market) AS avg_days_on_market
FROM property
JOIN sales
    ON property.listing_id = sales.Listing_ID
GROUP BY property.metro_distance_km;

"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#18-What is the monthly sales trend?
query = """
SELECT 
    DATE_FORMAT(sales.Date_Sold, '%%Y-%%m') AS month,
    COUNT(*) AS total_sales
FROM sales
GROUP BY DATE_FORMAT(sales.Date_Sold, '%%Y-%%m')
ORDER BY month;
"""

result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#19-Which properties are currently unsold?
query = """
SELECT *
FROM listings
WHERE Listing_ID NOT IN (
    SELECT Listing_ID
    FROM sales
)
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#20-Which agents have closed the most sales?
query = """
SELECT 
    a.Name,
    COUNT(*) AS total_properties
FROM listings l
JOIN agents a
ON l.Agent_ID = a.Agent_ID
GROUP BY a.Name
ORDER BY total_properties DESC
"""

result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#21-Who are the top agents by total sales revenue?
query = """
SELECT 
    a.Name,
    SUM(l.Price) AS Total_Price
FROM listings l
JOIN agents a
ON l.Agent_ID = a.Agent_ID
GROUP BY a.Name
ORDER BY Total_Price DESC;
"""

result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#22-Which agents close deals fastest?
query = """
SELECT 
    a.Name,
    AVG(s.Days_on_Market) AS Avg_Days_On_Market
FROM sales s
JOIN listings l
ON s.Listing_ID = l.Listing_ID
JOIN agents a
ON l.Agent_ID = a.Agent_ID
GROUP BY a.Name
ORDER BY Avg_Days_On_Market;
"""

result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#23-Does experience correlate with deals closed?
query = """
SELECT 
agents.experience_years,
deals_closed
FROM agents
ORDER BY experience_years DESC;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#24-Do agents with higher ratings close deals faster?
query = """
SELECT 
    agents.Agent_ID,
    agents.rating,
    AVG(sales.Days_on_Market) AS Avg_Days_On_Market
FROM sales
JOIN listings
ON sales.Listing_ID = listings.Listing_ID
JOIN agents
ON listings.Agent_ID = agents.Agent_ID
GROUP BY agents.Agent_ID, agents.rating;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#25 - What is the average commission earned by each agent?
query = """
SELECT 
    Agent_ID,commission_rate
FROM agents;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#26-Which agents currently have the most active listings?
query = """
SELECT 
Agent_ID,
    COUNT(*) AS total_listings
FROM listings
GROUP BY Agent_ID;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#27-What percentage of buyers are investors vs end users?
query = """
SELECT 
    buyer_type,
    COUNT(*) * 100.0 / (SELECT COUNT(*) FROM buyers) AS Percentage
FROM buyers
GROUP BY buyer_type;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#28-Which cities have the highest loan uptake rate?
query = """
SELECT 
    listings.City,
    AVG(buyers.loan_taken) * 100 AS Loan_Uptake_Percentage
FROM buyers
JOIN listings
ON buyers.sale_id = listings.Listing_ID
GROUP BY listings.City
ORDER BY Loan_Uptake_Percentage DESC;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#29-What is the average loan amount by buyer type?

query = """
SELECT 
    buyer_type,
    AVG(loan_amount) AS Avg_Loan_Amount
FROM buyers
GROUP BY buyer_type;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#30-Which payment mode is most commonly used?
query = """
SELECT 
    payment_mode,
    COUNT(*) AS Count
FROM buyers
GROUP BY payment_mode
ORDER BY Count DESC;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```

```python
#31-Do loan-backed purchases take longer to close?
query = """
SELECT 
    buyers.loan_taken,
    AVG(sales.Days_on_Market) AS Avg_Days_On_Market
FROM buyers
JOIN sales
ON buyers.sale_id = sales.Listing_ID
GROUP BY buyers.loan_taken;
"""
result_df = pd.read_sql(query, engine)
result_df.head()
```
